```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,100)
x1 = np.linspace(-6, 6, 100)
x2 = np.linspace(0, 5, 100)
```

```python
y = x**2 + 5
fig = plt.figure()
plt.plot(x,y, label = "y = x^2 + 5")
plt.title("Wykres funkcji x^2 + 5")
plt.xlabel("Argument x")
plt.ylabel("WartoÅ›Ä‡ funkcji")
plt.legend()
plt.show()
```
