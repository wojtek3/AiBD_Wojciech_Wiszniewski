```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DIR = "../AnalysisData/"
df = pd.read_csv(f"{DIR}tb.csv")

df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[2:],
             var_name='column', value_name='cases')

df = df[df['cases'] == df['cases']]
df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[7] 
df['age'] = df['column'].str[8:].map({
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+',
    'u' : '65+'
})
df.to_csv(path_or_buf=f"{DIR}tb_melted.csv", index=False)
df
```

```python
#Zachorowania na gruźlicę w Polsce w każdej krupie wiekowej i każdym roku

dfPL = df[df['iso2'].isin(['PL'])]
dfPL = dfPL[~dfPL['column'].isin(['new_sp'])]
dfPL
```

```python
#Zachorowania na gruźlicę w USA w każdej krupie wiekowej i każdym roku

dfUS = df[df['iso2'].isin(['US'])]
dfUS = dfUS[~dfUS['column'].isin(['new_sp'])]
dfUS
```

```python
##Zachorowania na gruźlicę w Polsce sumarycznie w każdym roku

dfPL = df[df['iso2'].isin(['PL'])]
dfPL = dfPL[dfPL['column'].isin(['new_sp'])]
dfPL
```

```python
##Zachorowania na gruźlicę w USA sumarycznie w każdym roku

dfUS = df[df['iso2'].isin(['US'])]
dfUS = dfUS[dfUS['column'].isin(['new_sp'])]
dfUS
```

```python
#zachorowania w Polsce w każdym roku
dfPL.plot(x="year", y="cases", kind="bar")
```

```python
#zachorowania w USA w każdym roku
dfUS.plot(x="year", y="cases", kind="bar")
```

```python
#zachorowania mężczyzn w Polsce w roku 2008 dla różnych grup wiekowych
dfPL = df[df['iso2'].isin(['PL'])]
dfPL = dfPL[~dfPL['column'].isin(['new_sp'])]
dfPL = dfPL[dfPL['year'].isin([2008])]
dfPL = dfPL[dfPL['sex'].isin(['m'])]
dfPL.plot(x="column", y="cases", kind="bar")
```

```python
#zachorowania kobiet w Polsce w roku 2008 dla różnych grup wiekowych
dfPL = df[df['iso2'].isin(['PL'])]
dfPL = dfPL[~dfPL['column'].isin(['new_sp'])]
dfPL = dfPL[dfPL['year'].isin([2008])]
dfPL = dfPL[dfPL['sex'].isin(['f'])]
dfPL.plot(x="column", y="cases", kind="bar")
```

```python
#zachorowania mężczyzn w USA w roku 2008 dla różnych grup wiekowych
dfUS = df[df['iso2'].isin(['US'])]
dfUS = dfUS[~dfUS['column'].isin(['new_sp'])]
dfUS = dfUS[dfUS['year'].isin([2008])]
dfUS = dfUS[dfUS['sex'].isin(['m'])]
dfUS.plot(x="column", y="cases", kind="bar")
```
