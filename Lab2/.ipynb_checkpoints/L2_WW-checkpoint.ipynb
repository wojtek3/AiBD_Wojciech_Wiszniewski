3. Znajdź najstarszy i najmłodszy film do wypożyczenia.

```python
df = pd.read_sql('select * from film order by release_year asc fetch first row only', con=connection_sqlalchemy)
df
```

4. Ile wypożyczeń odbyło się między 2005-07-01 a 2005-08-01?

```python
df = pd.read_sql("select * from rental where rental_date between '2005-07-01 00:00:00' and '2005-08-01 23:59:59'", con=connection_sqlalchemy)
len(df.index)
```

5. Ile wypożyczeń odbyło się między 2010-01-01 a 2011-02-01?

```python
df = pd.read_sql("select * from rental where rental_date between '2010-01-01 00:00:00' and '2011-02-01 23:59:59'", con=connection_sqlalchemy)
len(df.index)
```

6. Znajdź największą płatność wypożyczenia.

```python
df = pd.read_sql('select * from payment order by amount desc limit 1', con=connection_sqlalchemy)
df
```

7. Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.

9. Ilu pracowników mieszka w Argentynie lub Hiszpanii?

10. Jakie kategorie filmów zostały wypożyczone przez klientów?

```python
df = pd.read_sql("select distinct category from sales_by_film_category where total_sales > 0", con=connection_sqlalchemy)
df
```

11. Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.

```python
df = pd.read_sql("select distinct category from film_list where FID in (select film_id from inventory where inventory_id in (select inventory_id from rental where customer_id in (select id from customer_list where country = 'United States')))", con=connection_sqlalchemy)
df
```
